```julia
Q[:,1:3]  # Span same as columns of A' / rows of A
K = Q[:,4:5] # is orthogonal to Q[:,1:3] i.e., orthogonal to rows of A
```

```julia
A = randn(5,5)
b = randn(5)
```

```julia
x_qr = qr(A) \ b
x_lu = lu(A) \ b
```

```julia
x_lu - x_qr
```

```julia
A \ b # uses LU. Why?
```

```julia
n = 100
A = randn(n,n)
b = randn(n)

@btime qr(A);
@btime lu(A);
```

```julia

n = 100
A = randn(n,n)
b = randn(n)

x_lu = A \ b;
x_qr = qr(A) \ b;
```

```julia
norm((big.(A) \ b) - x_lu)
```

```julia
norm((big.(A) \ b) - x_qr)
```

```julia
function badmatrix(n)
    A = Matrix(1I, n, n)
    A[:,end] .= 1
    for j = 1:n-1
        A[j+1:end,j] .= -1
    end
    A
end
A = badmatrix(5)
```

```julia
lu(A)
```

```julia
n = 200
A = badmatrix(n)
b = randn(n)
norm(A \ b - qr(A) \ b) # lu fails on this matrix
```

```julia
A_ = A .+ 0.0001randn.();
```

```julia
norm(A_ \b - qr(A_)\b) # perturbed LU is fine
```
