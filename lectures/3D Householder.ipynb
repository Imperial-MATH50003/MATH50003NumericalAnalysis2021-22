```julia
using Plots, LinearAlgebra, Interact
```

```julia
@manipulate for x=1, y=-1, z=1
     = [x,y,z]
     = -norm()*[1,0,0] + 

    scatter3d([[1]], [[2]], [[3]]; label="x", legend=:bottomleft)
    scatter3d!([[1]], [[2]], [[3]]; label="y")
    plot!([0,[1x]], [0,[2]], [0,[3]]; arrow=true, label=nothing, linecolor=:blue)
    plot!([0,[1]], [0,[2]], [0,[3]]; arrow=true, label=nothing, linecolor=:red)

    xs, ys = [-2,2], [-2,2]

    Q = nullspace(')
    a,b = Q[1:2,:]' \ Q[3,:]
    p = (x,y) -> a*x + b*y

    surface!(xs, ys, p.(xs', ys); alpha=0.5)
end
```
